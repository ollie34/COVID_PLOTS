# Confirmed Covid Cases Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#read in from github
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')

#Format Data
confirmed = confirmed.drop(columns=['Province/State', 'Lat', 'Long'])
confirmed = confirmed.groupby('Country/Region').agg('sum')
confirmed = confirmed.T

confirmed_date = pd.to_datetime(confirmed.index)
datetime_index = pd.DatetimeIndex(confirmed_date.values)
confirmed = confirmed.set_index(datetime_index)

#COUNTRY SELECTION
confirmed = confirmed[['US', 'New Zealand', 'Brazil', 'Mexico', 'Russia', 'Germany']]

#Plotting
plot = confirmed.plot(figsize=(16,8), linewidth=5)
plot.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plot.set_title('Confirmed Covid Cases Over Time', fontsize = 30)
plot.set_xlabel('Date', fontsize = 14)
plot.set_ylabel('# of Cases', fontsize = 14)
```
